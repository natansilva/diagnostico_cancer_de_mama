```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
base = pd.read_csv('dados/base.txt', sep=';', decimal=',')
```

Transformando a variavel resposta em numero, quando for Maligno será 1 e quando for Benigno será 0

```python
base['diagnostico'] = base['diagnostico'].apply(lambda x: 1 if x == 'M' else 0)
```

```python
plt.figure(figsize=[20, 10])

plt.title('Correlação linear entre as variaveis')

sns.heatmap(base.drop(['ID'], axis=1).corr(), annot=True, fmt='.2f', vmin=-1, vmax=1, cmap=sns.diverging_palette(999, 999, n=20))
plt.show()
```

### Feature Engineering

```python
from sklearn.decomposition import PCA

pca_trafo = PCA().fit(base.drop(['ID'], axis=1))

plt.semilogy(pca_trafo.explained_variance_ratio_, '--o')
plt.show()
```

### Gerando uma base de teste e treino

```python
X = base.drop(['ID', 'diagnostico'], axis=1)
y = base['diagnostico']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Treinando um modelo de classificação Naive Bayes

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
accuracy = accuracy_score(clf.predict(X_test), y_test)
precision = precision_score(clf.predict(X_test), y_test)
recall = recall_score(clf.predict(X_test), y_test)

print('Acuracia: %.2f' % accuracy)
print('Precisão: %.2f' % precision)
print('Recall: %.2f' % recall)
```

### Treinando um modelo de classificação Regressão Logistica

```python
clf = LogisticRegression()
clf = clf.fit(X_train, y_train)
```

```python
accuracy = accuracy_score(clf.predict(X_test), y_test)
precision = precision_score(clf.predict(X_test), y_test)
recall = recall_score(clf.predict(X_test), y_test)

print('Acuracia: %.2f' % accuracy)
print('Precisão: %.2f' % precision)
print('Recall: %.2f' % recall)
```
